###### 2. Attenuation, Latency, and Jitter
***

After completing this topic, you should be able to: ``troubleshoot attenuation, latency and jitter given a scenario``

In this presentation we'll take a look at some of the factors that can contribute to poor signal quality, specifically over wired connections. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Attenuation
***

Beginning with attenuation, this refers to the reduction of signal strength while transmitting over long distances. 

And this is applicable to both analog and digital transmissions because in both cases you are still pushing an electrical signal down a physical wire. 

So sooner or later it will just run out of strength, it will not last forever, of course. 

So the amount of attenuation is measured in decibels. 

And, in fact, you might have something like a stereo amplifier at home that actually refers to the volume as attenuation. So it quite literally gets quieter, if you will, the farther it has to travel down that wire. 

Now you might see it also measured in terms of voltage. But again, sooner or later the signal strength will just degrade to the point where it is no longer detectable. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Attenuation - Transmission Medium
***

So in terms of the causes, the transmission medium itself is in essence the main cause. 

As you push an electrical signal down a metal wire it causes an electromagnetic field. And this in essence saps energy from the transmission of the packet itself and goes into creating the field. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Attenuation - Crosstalk
***

So this can also happen with adjacent cabling because, of course, they are doing the exact same thing your cable is doing. 

So they all generate these electromagnetic fields, and they can actually interfere with each other to a degree. 

And this is why you see things like shielded cabling to reduce that as much as possible.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Attenuation - Conductors
***

But also different conductors. The different types of materials that are used as a signal passes from, let's say, the physical wire, into something like a router or a switch or any other interconnectivity device. Or even from network to network where the material changes. 

So all of that can contribute to attenuation. 

But for the most part it is simply the resistance of the cable itself. Again, signals do not travel forever down a physical medium. 

So sooner or later, they just die out. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Latency
***

Latency is an expression of how long it takes for data to get from one point to another. 

Quite simply, the amount of time. And any given delay can typically be cumulative. 

In other words, a bottleneck in one area can certainly have an effect elsewhere because you need to wait until everything gets through in many cases before you can process something and move it on its way. 

So ideal conditions in a perfect world, latency should be as close to zero as possible.

Now, of course, you'll never actually see zero, it has to travel some distance therefore it must take some amount of time. 

But the closer you can get to zero, the better. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Latency - Propagation
***

So in terms of the causes of latency, propagation refers to the amount of time it takes for data to travel from source to destination. 

But in many cases it's source to multiple destinations, there are many instances where you send packets to more than one location. 

So it's sort of the cumulative effect, if you will. In other words it's the amount of time it takes for all packets to reach all destinations. That refers to the latency in total. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Latency - Transmission Medium
***

And, of course, the transmission medium itself is going to have an effect. 

Ethernet versus fiber will certainly vary in terms of the latency, because let's say copper wire Ethernet typically maxes out at about 100 meters. 

Therefore you have to pass the signal through many more interconnection devices to traverse a long distance. 

Whereas fiber can travel for kilometers before it ever needs to be boosted or passed through any kind of interconnection device. 

So it's just all those stops, if you will, that an Ethernet environment has to implement that will quite simply delay the delivery of the packet.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Latency - Network Devices
***

Network devices such as routers also will increase the latency because they have to examine the packets and make a decision as to where it needs to go.

And it also has to increment the hop count. Every time it passes a router, it increments that hop count to in essence say it passed over another router. 

Distance vector routing protocols use that hop count. So they have to process the packet. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Latency - Hard Disk Access
***

And hard disk access delays, of course, just simply retrieving the data in the first place, will take some time. 

Then how much does it have to be switched? Which bridges does it have to cross? Things along those lines. So again, just more processing has to come into the picture. 

So the more processing that's involved, the longer it will take. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Jitter
***

And jitter refers to deviation in a signal pulse with respect to its amplitude, its phase timing and or the width of the signal pulse. 

And there can also be, of course, the variation in latency due to the fact that some packets take longer to arrive at any given system than do others. 

So all of this can result in jitter. 

And network congestion, timing drift, and route changes can certainly contribute to jitter. 

And this is something that is typically a problem in real-time communication such as IP telephony, or voice over IP, or video conferencing. That's where you notice it the most.

Because if a packet arrives out of sequence or just in the wrong order, then this has to be reassembled and causes a delay. So that's when you just really see it the most. 

When it's something that's not really concerned with real-time such as transferring a file, you don't ever really see it. It might still take a little more time but it's not something you notice, the file still comes up in its exact correct state.

But real-time communications ideally would like to see the least amount of latency and the least amount of jitter, so that the communication goes as smoothly as possible.

###### 3. Crosstalk and EMI
***

After completing this topic, you should be able to: ``given a scenario troubleshoot crosstalk and EMI``

In this presentation, we'll take a look at crosstalk and electromagnetic interference.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Crosstalk
***

Now crosstalk occurs when signals from one wire interfere with signals on another. 

Now this is a more specific symptom of the larger problem, if you will.

In other words, it is caused by electromagnetic interference. But it quite simply refers to the fact that any given electric signal can cross from one wire to another when they are in very close proximity to each other. 

In terms of measuring it, this is a matter of perspective, if you will. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Crosstalk - Near-end (NEXT)
***

In other words, if I'm at one end of any given connection and you are at the other, then it's sort of a matter of well, is the problem on my end or your end? 

So this obviously depends on the perspective, but near-end crosstalk, or NEXT, is found close to the connector. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Crosstalk - Far-end (FEXT)
***

And far-end crosstalk, or FEXT, is similar but at the far end. 

So again, from my prospective, is it on my end or your end? 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Complications - Error (Noise)
***

So the complications that can arise of course include really any kind of error and noise, quite literally. 

Now in terms of noise, you might actually remember back in days of purely analog phone calls. You could actually hear someone on another line from time to time. This was due to crosstalk. 

Now it's not as big a deal anymore with digital communications and quite simply better isolation. 

But you literally could hear other people in those scenarios. 

Now in terms of digital communication such as transferring a file, we don't really think of it as something that you can hear. But it is still interference, so we call it noise. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Complications - Packet Loss
***

Other complications include packet loss. 

Now it's not the fact that a packet might entirely disappear. It might still arrive, but some of the data is no longer intact. 

So there's a certain amount of loss of data and flat out transmission failure. Where a packet just does not arrive at all, and then of course, you end up having to retransmit. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Prevention - Twisted Pair Cable or better yet Shielded Cable
***

So in terms of prevention, you can implement twisted pair cable, or better yet, shielded cable. 

But even if it is unshielded, the twisted pair cabling does provide a little bit of extra insulation, because you are twisting the pairs of cables around each other. 

So that does give you that little bit of extra insulation. 

But shielded cable, of course, will work better. 

But if it's too bad, then you might need to physically separate those cables. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Electromagnetic Interference
***

So as mentioned, crosstalk is a little bit of a more specific symptom of a larger problem, electromagnetic interference. 

And this is any kind of interference caused by any kind of electromagnetic disturbance. 

And this can affect the overall performance of really any type of device or any type of data transmission. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Electromagnetic Interference - Radio Frequency Interference
***

Now closely related to that is also radio frequency interference. 

And this of course is any kind of interference that is occurring on a radio frequency spectrum, such as two devices trying to transmit on the same frequency to the same device. 

It would be like trying to tune your radio to two channels at the same time. Clearly they would just interfere with each other, but radio frequency is still a type of electromagnetic transmission. 

So those two are closely related. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Electromagnetic Interference - Conducted EMI or Induction
***

Now this of course can be caused by direct physical contact with a conductor.

And this is referred to conducted EMI or induction, where there's no physical contact but rather radiated EMI. 

Being too close to an electromagnetic field, such as electric power cords or unshielded cabling. 

Now this is commonly observed with radios and as mentioned, this is sort the static noise heard when switching between channels. But if it was possible to tune into two channels at the same time, it would still be fairly illegible. You wouldn't be able to understand what's happening. 

And even with the TV, the picture can become distorted when the signal weakens, due to the interference. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Conclusion - Avoid
***

So ideally, what you want to try to do is just avoid it in the first place, so keep away from heavy machinery, any kind of motor, generator. 

Use shielded cables when necessary, so in ethernet, for example, STP, shielded twisted pair. 

And if necessary, some sort of EMI filter. Something that can quite literally shield your cables, for example, from the EMI, so that you can just have the best conditions possible in terms of communication from point A to point B.

###### 4. Opens and Shorts
***

After completing this topic, you should be able to: ``given a scenario troubleshoot opens or cable shorts``

In this presentation, we'll take a look at two possible problems with wired connections known as shorts and opens. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Short
***

Now, a short quite simply refers to a short circuit. 

And this is generally due to a section of a wire that has maybe become exposed and has come into contact with something else that can carry any kind of current. 

So if for example, we are talking about electrical cabling, then quite literally, the cables might have come into contact with a metal frame, or even with each other. And this additional point of contact has provided an alternate path at very low impedance or resistance for that current to follow. 

So because of this at the physical level, the circuit can overheat. You get a build up of too much power at that point, which clearly can damage the components and even cause a fire. Again, if you're talking about flat out power cabling. 

Now when it comes to networking problems, an intermittent short can certainly still cause network issues, such as broadcast storms or intermittent periods of no network traffic. Now this could be caused by something such as a faulty network adapter or even an entire switch. Or maybe there's issues with the power itself. If it's spiking, for example, that can actually cause an arc whereby a current is quite literally able to jump from one place to another when there is no actual physical contact. Now, it's usually when they're very close, but it's something that of course might go away on its own. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Continuous Short
***

But if you have a continuous short, then this will cause continuous packet collisions and continuous broadcast storms because we are literally talking about now physical wires touching each other.

So in that event it will always result in a collision and this generally means that something is flat out broken and needs to be replaced. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Test - Low-voltage - High-voltage
***

So there are two common ways to test for a short and is quite simply the voltage. 

Now it's either a low voltage test or a high voltage test depending on what it is you're dealing with. 

In other words if you are talking about electrical cabling, then that is a relatively high voltage. 

Your standard power, for example in your house or your office, runs up around 120 volts. 

So relatively speaking, in terms of network communications, that's very high voltage. 

But a packet running around on a network wire is much, much lower in terms of the voltage so it depends on what you are testing. 

But if it is fairly high in terms of its voltage, so again power, you might actually be able to visually identify where it's happening. Because if it's exposed and in contact with something else, you can very likely see sparks coming from that point. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Open
***

So an open is the effective opposite of a short. 

An open is when a cable connection is not continuous to the intended termination. 

In other words it might simply be disconnected or maybe something has cut the cable. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Test - Continuity Test - Resistance Test
***

So there are two ways to test for an open. 

A continuity test, meaning is it continuous, or a resistance test. Do you see the correct amount of resistance? Because again, every electrical type of cabling, or any cable that is carrying an electric signal offers some kind of resistance. And it should be consistent throughout the entire cable.

In other words the resistance should not vary. So, if I take a cable and cut it in half and separate those two ends, that is in essence infinite resistance. There's no way it can push that current across the break. 

So again, you can test for either and if it's, again, an open, clearly this is going to prevent the current from flowing. So any kind of network communication is clearly going to be ceased. 

Now you can, again, try to track these down by using a toner probe, which sends a fairly strong electromagnetic current down the cable. And you can try to pick up on that, on the other end, or you can also just touch the two ends of the cable to the probe. And if you don't hear a tone, then the cable has an open somewhere. 

In other words, it is simply not getting from one end all the way to the other end, and in many cases, it's just the cable has become disconnected. Somebody might have tripped over it and pulled it out. It's a matter of simply plugging it back in. 

But if, of course, they damaged the cable when they pulled it out, then you might need to replace the cable, but those of course are fairly easy fixes. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Conclusion - Electrical Problems - Replace - Ultimately try to identify and correct Shorts and Opens
***

If you are talking about electrical problems within a device itself, you might be looking at needing to replace the whole device. But ultimately you want to try to identify and correct shorts and opens as quickly as possible.

###### 6. Duplex and Transceiver Mismatches
***

After completing this topic, you should be able to: ``troubleshoot duplex and transceiver mismatches given a scenario``

Another communications error that can arise is one known as a duplex mismatch.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Half Duplex
***

Now in terms of just communicating, there is a method known as half duplex, whereby you use a single channel of communication.

Now you can both send and receive on that channel, but it has to be one at a time. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Full Duplex
***

Full duplex, which is implemented in technologies such as 10BaseT, Ethernet switches, and some wide area networking technologies, uses separate channels or wires.

And as such you can send and receive at the same time. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Duplex Mismatch - Auto Negotiation - 802.3u
***

So if you have devices that are capable of doing either, they need to be matched. 

And this was defined as what's known as auto negotiation. This was an implementation known as the IEEE 802.3u standard. 

And this eliminates the need to manually configure settings on each network device. 

The devices quite literally choose the common transmission parameters between them including the speed, the duplex mode and the flow. 

So this eliminates a lot of problems in terms of a duplex mismatch, they quite literally negotiate between themselves which properties need to be used. 

But if you do end up with a full duplex device trying to communicate with a half duplex device, this can result in the half duplex device stopping transmission and retransmitting packets at a later time. And it might list the interruption as a collision, or what's also known as a late collision error. But, in essence, it just doesn't understand, for a lack of a better word. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Identifying Mismatch
***

So identifying mismatches can be difficult. 

So to be certain you can always configure the devices manually. 

Most network adaptors have these properties that you can set but they're usually set to auto by default. 

So it's more time consuming, of course, to configure them manually, but you can ensure that the links are operating at the desired speed and or the correct duplex setting. So it might take a little longer but it might also be a little more reliable. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Transceiver Mismatch - Medium Access Unit (MAU)
***

Now the transceiver, or what's often referred to as the MAU, or Medium Access Unit, is the actual component, the actual device used to send and receive signals. 

And that goes for both analog and digital communications. 

And in most cases the transceiver is built right into the network interface card for just about every modern adaptor. But in some cases depending on the age and or what the device is, the transceiver might be external, in which case that's where you need to ensure the correct properties are set. 

So that whatever this transceiver is communicating with on the other end simply has the same configuration. 

So really what you are looking to implement is consistency, so that you can avoid these mismatch problems. 

And again, it's not that much of an issue these days for most adaptors and most devices, they will auto negotiate correctly. 

But it is something to be on the lookout for if you have a mix of equipment, particularly older and newer.

In this presentation we'll talk about some of the issues that can result from a damaged cable.

###### LED Status Indicators
***

But it does actually give you a means to troubleshoot in the form of LED status indicators. 

There is usually a status LED on virtually every network interface, and it will often have one of three states, solid, flashing or off. 

Now there's a little bit of variance there to a degree and I'll come back to that in a moment. 

But usually, a solid green light indicates that connectivity is present. So if, for example, you have a laptop and you plug the cable into the network interface, it should light up. And that indicates that it has successfully connected to something on the other end. If it does not, then there's something wrong with the cable or the connection on the other end, or possibly, of course, your own interface. But in essence, something is not working. 

Now the flashing is a little bit of a relative term, because if you do plug it in and you get a solid light, as soon as traffic starts to pass over that interface, you'll usually see very rapid flashing, more like blinking. But if it's a very slow flash, like maybe once every second or half a second, then that usually indicates some kind of problem. And we'll see an example in moment. 

So as mentioned, the flashing green is possibly a sign of intermittent connectivity. So this would be a fairly slow flash, again, maybe once every second or half a second. And that might indicate that either the network interface card or the Ethernet cable is going bad. There's some kind of problem that's not entirely gone but on its way out. But as I said, if you do have a solid stable connection then very often the flashing is very, very rapid. Looks more so like it's vibrating almost, or jittering if you will. 

So flashing, again, is a little bit of a relative term, but this would be a fairly slow, very easily detectable flash. 

You might also see a flashing amber, and this is generally an indication that there are collisions occurring on the network. 

And solid amber is often an indication that there is a fault with the card itself. 

If it's alternating green and amber very rapidly then the card might just be booting up. And if it's fairly slow, then the card might just be making DHCP requests, which in some cases might take a few minutes. But that ultimately should settle down and you should see, again, that solid green. 

And once traffic starts to flow again, you might see that very rapid jittery flashing, but generally that's what you want to see.
